```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread('/content/selfie.jpg')

# Converter a imagem de BGR para RGB
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Exibir a imagem
plt.imshow(imagem_rgb)
plt.axis('off')
plt.show()
```

```python
# Converter para grayscale
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Aplicar suavização com filtro Gaussiano
imagem_suavizada = cv2.GaussianBlur(imagem_gray, (15, 15), 0)
imagem_suavizada_rgb = cv2.cvtColor(imagem_suavizada, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_suavizada_rgb, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Detectar bordas
bordas = cv2.Canny(imagem_gray, 100, 200)
plt.imshow(bordas, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# desenhar retangulo na imagem
inicio = (50, 5)
fim = (220, 160)
cor = (255, 0, 0)
espessura = 2

# desenhar retangulo na copia da imagem
imagem_com_retangulo = cv2.rectangle(imagem.copy(), inicio, fim, cor, espessura)

# converter para rgb
imagem_com_retangulo_rgb = cv2.cvtColor(imagem_com_retangulo, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_com_retangulo_rgb)
plt.axis('off')
plt.show()
```

```python
# salar a imagem processada
cv2.imwrite('imagem_processada.jpg', imagem_com_retangulo)
```
