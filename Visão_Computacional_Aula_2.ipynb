```python
#from google.colab import file

# upload de arquivo
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]
```

```python
from paddleocr import PaddleOCR

# Inicializa o objeto PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Realiza a OCR na imagem
#result = ocr.ocr(pdf_path)
```

```python
# Função para extrair texto OCR de um PDF ou imagem
def extract_text_from_pdf(pdf_path):
  extract_ocr = ocr.ocr(pdf_path)

  result = ''
  for line in extract_ocr:
    for word in line:
      result += word[1][0] + ' '
    result += '\n'
  return result

# Extrai texto
ocr_text = extract_text_from_pdf(pdf_path)
print("Texto extraido")
print(ocr_text)
```

```python
# Extrair CNPJ
import re

# Expressão regular
padrao_cpf_cnpj = r'(CNPJ: \d{2}\.\d{3}\.\d{3}\/\d{4}\-\d{2})'
```

```python
# Função para Extrair informações
def extrair_informacoes(texto, padrao):
  match = re.search(padrao, texto)
  if match:
    return match.group(1)
  else:
    return None

# Texto
print(ocr_text)

# Extrair informações
cnpj = extrair_informacoes(ocr_text, padrao_cpf_cnpj)
print(cnpj)

# Verificar se há cnpj
cnpj_presente = re.search(padrao_cpf_cnpj, ocr_text) is not None
print("CNPJ presente?", bool(cnpj_presente))
```
